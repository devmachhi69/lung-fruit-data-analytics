```python
import pandas as pd


df = pd.read_csv('C:/Users/student/Downloads/fruit-data.csv')


print("Dataset Preview:")
print(df.head())


print("\nDataset Info:")
print(df.info())
```

```python
print("\nMissing values in 'mass':", df['mass'].isnull().sum())


df['mass'].fillna(df['mass'].mean(), inplace=True)
```

```python
df['fruit_name'] = df['fruit_name'].str.upper()
print("\nUpdated fruit_name column:")
print(df['fruit_name'].head())
```

```python
duplicates = df[df.duplicated()]
print("\nDuplicate Records:")
print(duplicates)
```

```python
df_backup = df.copy()  # backup

df.drop(columns=['fruit_subtype'], inplace=True)

print("\nData after dropping 'fruit_subtype':")
print(df.head())
```

```python
print(df.columns)

```

```python
import matplotlib.pyplot as plt
avg_fruit_label = df.groupby('mass')['fruit_label'].mean()

avg_fruit_label.plot(kind='bar', color='orange')
plt.title('Average Fruit Label by Mass')
plt.xlabel('Mass')
plt.ylabel('Average Fruit Label')
plt.show()
```

```python

height_counts = df['height'].value_counts()
plt.pie(height_counts, labels=height_counts.index, autopct='%1.1f%%', colors=plt.cm.Pastel1.colors)
plt.title('Proportion of Height')
plt.show()

```

```python

df_sorted = df.sort_values('width')
plt.plot(df_sorted['width'], df_sorted['fruit_label'], marker='o')
plt.title('Fruit Label Trend by Width')
plt.xlabel('Width')
plt.ylabel('Fruit Label')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt


height_counts = df['height'].value_counts()


plt.figure(figsize=(7,7))
plt.pie(
    height_counts,
    labels=height_counts.index,
    autopct='%1.1f%%',
    colors=plt.cm.Pastel1.colors,
    startangle=90,
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Proportion of Height')
plt.axis('equal')  
plt.show()
```
